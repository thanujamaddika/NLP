```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/suhasmaddali/Twitter-Sentiment-Analysis/refs/heads/main/train.csv')
```

```python
df = df[['selected_text', 'sentiment']]
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt_tab')
nltk.download('stopwords')
import re

def preprocess_text(text):
    if isinstance(text, float):
        return ""
    text = text.lower()  # Lowercase
    text = re.sub(r'http\S+|www.\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-zA-Z ]', '', text)  # Remove special characters & numbers
    words = nltk.word_tokenize(text)  # Tokenization
    words = [word for word in words if word not in stopwords.words('english')]  # Remove stopwords
    return ' '.join(words)

tokens = []
# Change 'text' to 'selected_text' to access the existing column
for i in range(len(df)):
    tokens.append(preprocess_text(df['selected_text'][i]))
```

```python
tokens
```

```python
from gensim.models import Word2Vec
```

```python
sentences=[word_tokenize(i) for i in tokens]
```

```python

model=Word2Vec(sentences,vector_size=100,window=5,min_count=1,workers=4)
```
